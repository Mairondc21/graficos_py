```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
df = pd.DataFrame(vendas_2022, index=lojas)
df
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,8))
fig.subplots_adjust(wspace=0.3, hspace=0.4)
fig.suptitle('Vendas de 2022')

axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Loja A')

axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Loja B')

axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Loja C')

axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Loja D')

for ax in axs.ravel():
    ax.set_xlabel('mês')
    ax.set_ylabel('Número de vendas')
    ax.set_ylim(0,500)



plt.show()
```
