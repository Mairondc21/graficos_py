```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País',inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brasil',anos]
```

```python
brasil
```

```python
brasil_dict = {'Ano':brasil.index.tolist(), 'Imigrantes':brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['Ano'],dados_brasil['Imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('anos')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['Ano'],dados_brasil['Imigrantes'], lw=3, color='b')
ax.set_title('Imigracao do Brasil para o Canadá\n1980 a 2013', fontsize=16)

ax.set_xlabel('Ano',fontsize=14)
ax.set_ylabel('Numero de Imigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
america_sul_ordenado = america_sul.sort_values('Total',ascending=True)
```

```python
cores = []
for pais in america_sul_ordenado.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')    

fig, ax = plt.subplots(figsize=(8,5))
ax.barh(america_sul_ordenado.index,america_sul_ordenado['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', fontsize=14)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))

axs[0].plot(dados_brasil['Ano'],dados_brasil['Imigrantes'])
axs[0].set_title('Imigracao do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['Imigrantes'])
axs[1].set_title('Boxplot da Imigracao do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numero de Imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores paises da America do Sul de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel('Ano')
    ax.set_ylabel('Numero de Imigrantes')

ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

plt.show()
```
